```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, classification_report

# Load dataset
# (Use your CSV file with columns like: label, message)
# Make sure 'spam.csv' is uploaded to your Colab environment or provide the correct path.
data = pd.read_csv('spam.csv', encoding='latin-1')

# Keep only useful columns (common in SMS Spam datasets)
data = data[['v1', 'v2']]
data.columns = ['label', 'message']

print(data.head())

# Convert labels: spam=1, ham=0
data['label'] = data['label'].map({'spam': 1, 'ham': 0})

# Split data
X_train, X_test, y_train, y_test = train_test_split(
    data['message'],
    data['label'],
    test_size=0.2,
    random_state=42
)

# Convert text to numbers
vectorizer = CountVectorizer()
X_train_vec = vectorizer.fit_transform(X_train)
X_test_vec = vectorizer.transform(X_test)

# Model
model = MultinomialNB()
model.fit(X_train_vec, y_train)

# Predictions
y_pred = model.predict(X_test_vec)

# Accuracy
print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))

# Test with custom message
msg = ["Congratulations! You won a free iPhone. Click here now!"]
msg_vec = vectorizer.transform(msg)

print("Prediction:", "Spam" if model.predict(msg_vec)[0] == 1 else "Ham")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Visualize the distribution of ham vs spam messages
plt.figure(figsize=(8, 6))
sns.countplot(x='label', data=data)
plt.title('Distribution of Ham (0) vs Spam (1) Messages')
plt.xlabel('Message Type')
plt.ylabel('Count')
plt.xticks(ticks=[0, 1], labels=['Ham', 'Spam'])
plt.show()
```
